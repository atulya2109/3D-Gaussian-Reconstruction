```python
import torch
import torchvision
import os
```

```python
frames_dir = '/home/atulyabist21/data/images/'
transform = torchvision.transforms.Compose([
    torchvision.transforms.Pad((0, 0, 0, 0), fill=0, padding_mode='constant'),
    torchvision.transforms.Resize((256, 256))
])

for filename in os.listdir(frames_dir):
    i = 0
    if filename.endswith(".jpg") or filename.endswith(".png"):
        i+=1
        print(i)
        img_path = os.path.join(frames_dir, filename)
        print(img_path)
        img = torchvision.io.read_image(img_path)
        width, height = img.size
        padding = (0, 0, max(height - width, 0), max(width - height, 0))
        padded_img = torchvision.transforms.functional.pad(img, padding, fill=0, padding_mode='constant')
        resized_img = transform(padded_img)
        resized_img.save(img_path)
```
